```python
# following scripts crawls all the iphone data available in flipkart website
from bs4 import BeautifulSoup
import requests

URL = 'https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'

content = requests.get(URL)

soup = BeautifulSoup(content.text, 'lxml')

iphones = []

for iphone in soup.findAll('div', class_='_1UoZlX'):

    name = iphone.a.img["alt"]

    price_tag = iphone.find('div', class_='col col-5-12 _2o7WAb')
    price = price_tag.div.div.div.text
    
    rating_tag = iphone.find('div', class_='niH0FQ')
    rating = rating_tag.div.text
    
    iphones.append([name, price, rating])

    
for iphone in iphones:
    print(iphone)

```

```python
import pandas as pd
df = pd.DataFrame(iphones, columns = ['Model', 'Cost', "Rating"])
df
```
